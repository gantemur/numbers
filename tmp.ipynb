```python
def harm(n):
    s = 1
    for i in range(2,int(n)+1):
        s += 1./i
    return s
```

```python
m=10
n=500
xres=np.linspace(m,n,n-m+1)
plt.figure(figsize=(15,8))
plt.plot(xres,[harm(x)-np.log(x) for x in xres])
plt.plot(xres,[0.577215664901532 for x in xres],'--')
#plt.savefig("eulergamma.png", dpi=200)
plt.show()
```

```python
def logfact(n):
    s = 0
    for i in range(2,int(n)+1):
        s += math.log(i)
    return s
```

```python
m=10
n=500
xres=np.linspace(m,n,n-m+1)
plt.figure(figsize=(15,8))
plt.plot(xres,[logfact(x)+x-(x+.5)*math.log(x) for x in xres],label='f(n)')
plt.plot(xres,[.5*math.log(2*math.pi) for x in xres],'--',label='0.5ln(2pi)')
plt.legend()
#plt.savefig("stirlingconstant.png", dpi=200)
plt.show()
```

```python
def fleft(x):
    return 2*x*math.log(2)

def fright(x):
    return 3*math.log(2*x)*(1+math.sqrt(2*x))
```

```python
def ft1(x):
    return x**(1./(x-1))

def ft2(x):
    a=ft1(x)
    return x*a*(2-a)
```

```python
a=.5
l2=math.log(2)
factor=2*(l2-a)
const=2*a+l2
def fff(x):
    return factor*x-math.log(x)-const
```

```python
def numdivs(n):
    if n == 1:
        return 1
    d = 2
    nn = int(math.sqrt(n))
    if nn*nn == n:
        d += 1
        nn -=1
    for i in range(2,nn+1):
        if i*int(n/i) == n:
            d += 2
    return d
```

```python
n=250
d = [numdivs(i) for i in range(1,n+1)]
xres=np.linspace(1,n,500)
plt.figure(figsize=(15,10))
plt.plot(xres,[d[int(x)-1] for x in xres])
#plt.legend()
#plt.savefig("numdivs.png", dpi=150)
plt.show()
```

```python
kk=0.577215664901532*2-1
n=30
d = [numdivs(i) for i in range(1,n+1)]
for i in range(len(d)-1):
    d[i+1] += d[i]
xres=np.linspace(1,n,1000)
plt.figure(figsize=(15,10))
plt.plot(xres,[d[int(x)-1] for x in xres],label='$D(x)$')
plt.plot(xres,[x*(math.log(x)+kk) for x in xres],label='$x\ln x-(2\gamma-1)x$')
plt.legend()
#plt.savefig("snumdivs.png", dpi=200, bbox_inches='tight')
plt.show()
```

```python
aa1 = [1]
bb1 =[2,2]

aa2 = [1]
bb2 = [2,3,6]

aa3 = [1,6]
bb3 = [2,3,3]

#Hammond
aa4 = [1,12]
bb4 = [2,3,4]

#Sylvester
aa5 = [1,6,70]
bb5 = [2,3,5,7,210]

aa6 = [1,15]
bb6 = [2,3,5,30]

aa7 = [1,6,10,210,231,1155]
bb7 = [2,3,5,7,11,105]

aa8=[1, 6, 10, 14, 105]
bb8=[2, 3, 5, 7, 11, 13, 385, 1001]

#Chebyshev
aac = [1,30]
bbc = [2,3,5]

#experiments
aax = [1,6,70]
bbx = [2,3,5,7,210]

def ECheb(x,aa,bb):
    r = 0
    for a in aa:
        r += int(x/a)
    for b in bb:
        r -= int(x/b)
    return r

def stig(aa,bb):
    r = 0
    for a in aa:
        r -= math.log(a)/a
    for b in bb:
        r += math.log(b)/b
    return r

def bstig(aa,bb,n):
    return stig(aa,bb)*n/(n-1)

def astig(aa,bb,n,m):
    return stig(aa,bb)*(1-n/(m*(n-1.)))
```

```python
def solve22(a,b,c,d,e,f):
    r = []
    r.append(a*e-d*b)
    r.append(c*e-f*b)
    r.append(a*f-d*c)
    return r

def eigval22(a,b,c,d):
    p=a+d
    q=a*d-c*b
    r=[]
    xx=p/2
    D=xx*xx-q
    if D>=0:
        DD=math.sqrt(D)
        r.append(xx-DD)
        r.append(xx+DD)
    return r

def iter2(a,b,c,d,e,f,x,y,n):
    xx = [x]
    yy = [y]
    for i in range(n):
        xn = a*x+b*y+c
        yn = d*x+e*y+f
        xx.append(xn)
        yy.append(yn)
        x=xn
        y=yn
    return xx,yy
```
